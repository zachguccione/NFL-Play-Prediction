```python
import pandas as pd
```

```python
df = pd.read_csv("../data/tracking_week_1.csv")
df.head()
```

```python
df = df[df['displayName'] == 'football']
df.head()
```

```python
# Filter for rows where frameId is 1 or the maximum frameId for each gameId and playId
filtered_df = df.groupby(['gameId', 'playId']).apply(
    lambda group: group[(group['frameId'] == 1) | (group['frameId'] == group['frameId'].max())]
).reset_index(drop=True)

filtered_df.head()
```

```python
# Group by `gameId` and `playId` to calculate the difference in y values between the first and last frames
play_summary = (
    df.groupby(['gameId', 'playId'])
    .apply(lambda group: pd.Series({
        'playDirection': group['playDirection'].iloc[0],
        'gameId': group['gameId'].iloc[0],
        'playId': group['playId'].iloc[0],
        'y_difference': group.loc[group['frameId'].idxmax(), 'y'] - group.loc[group['frameId'].idxmin(), 'y']
    }))
    .reset_index(drop=True)
)

play_summary
```

```python
# Export the dataframe to a CSV file
play_summary.to_csv("../data/play_direction.csv", index=False)
```
